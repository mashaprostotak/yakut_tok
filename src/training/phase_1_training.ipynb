```python
from transformers import Trainer, TrainingArguments, DataCollatorForLanguageModeling
from datasets import Dataset
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
import os
from google.colab import drive
import glob

# Mount Google Drive
drive.mount('/content/drive')

# Check GPU availability
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nUsing device: {device}")
if device.type == 'cuda':
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")

# 1. Define paths
model_path = "/content/drive/MyDrive/yakut-llama-model2"  # Path to your custom model
data_dir = "/content/drive/MyDrive/Yakut_Raw_Data"  # Directory with Yakut text files

# 2. Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path).to(device)  

# 3. Configure tokenizer padding
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.pad_token_id

# 4. Prepare dataset 
def load_yakut_corpus(directory):
    """Load and process all Yakut text files in directory"""
    texts = []
    # Get all text files in directory
    file_paths = glob.glob(os.path.join(directory, "*.txt"))

    if not file_paths:
        raise ValueError(f"No text files found in directory: {directory}")

    print(f"Found {len(file_paths)} text files")

    for file_path in file_paths:
        print(f"Processing: {os.path.basename(file_path)}")
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
            for line in f:
                text = line.strip()
                if len(text) > 10:  # Skip very short lines
                    # Format with special tokens
                    texts.append(f"<|begin_of_text|><sah>{text}<|end_of_text|>")
    return texts

print("\nLoading Yakut corpus...")
all_texts = load_yakut_corpus(data_dir)
print(f"Loaded {len(all_texts)} text samples")

# Create dataset
dataset = Dataset.from_dict({"text": all_texts})
train_test_split = dataset.train_test_split(test_size=0.05)
train_dataset = train_test_split["train"]
eval_dataset = train_test_split["test"]

# 5. Tokenization function
def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        max_length=512,
        padding=False,
        return_overflowing_tokens=False,
    )

print("\nTokenizing dataset...")
tokenized_train = train_dataset.map(
    tokenize_function,
    batched=True,
    num_proc=4,
    remove_columns=["text"],
    desc="Tokenizing training data"
)

tokenized_eval = eval_dataset.map(
    tokenize_function,
    batched=True,
    num_proc=4,
    remove_columns=["text"],
    desc="Tokenizing evaluation data"
)

# 6. Data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,
)

# 7. Training configuration 
training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/yakut-llama2-finetuned",
    overwrite_output_dir=True,
    num_train_epochs=5,               
    per_device_train_batch_size=8,     
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=32,     
    learning_rate=3e-5,               
    weight_decay=0.01,
    warmup_ratio=0.1,                  
    logging_steps=100,
    eval_steps=500,                    
    save_steps=1000,
    save_total_limit=2,
    fp16=True,                         # Force FP16 for GPU efficiency
    report_to="none",
    logging_dir="/content/drive/MyDrive/yakut-finetuned-logs",
    dataloader_num_workers=4,         
    gradient_checkpointing=True,       
    optim="adafactor",                 
    eval_strategy="steps",
)

# 8. Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_eval,
)

# 9. Start training
print("\nStarting fine-tuning...")
trainer.train()

# 10. Save final model
final_model_path = "/content/drive/MyDrive/yakut-llama2-finetuned"
trainer.save_model(final_model_path)
tokenizer.save_pretrained(final_model_path)
print(f"\nFine-tuned model saved to: {final_model_path}")

# 11. Test the fine-tuned model
def generate_text(prompt, model, tokenizer, max_length=50):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    outputs = model.generate(
        **inputs,
        max_length=max_length,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.8,
        repetition_penalty=1.1,
        pad_token_id=tokenizer.pad_token_id
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=False)

print("\nTesting fine-tuned model:")
model.eval()  # Set to evaluation mode

test_prompts = [
    "<|begin_of_text|><sah>Мин аатым ",
    "<|begin_of_text|><sah>Саха сирэ ",
    "<|begin_of_text|><sah>Күн бүгүн ",
    "<|begin_of_text|><sah>Оҕолор ",
    "<|begin_of_text|><sah>Биһиги ",
    "<|begin_of_text|><sah>Тыыннаах айылҕа ",
    "<|begin_of_text|><sah>Бииргэ үлэлии ",
]

for prompt in test_prompts:
    with torch.no_grad():
        generated = generate_text(prompt, model, tokenizer)
        print(f"\nPrompt: {prompt}")
        print(f"Generated: {generated}")
        print("-" * 80)

# Clean up GPU memory
torch.cuda.empty_cache()
print("GPU memory cleared")
```
